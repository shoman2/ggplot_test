```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&oid=028&aid=0002439215&iid=3674552&sid1=110'
```

```python
news = requests.get(url).text
```

```python
parser = BeautifulSoup(news, 'html.parser')
```

```python
parser.select('#articleBodyContents')
```

```python
for tag_text in parser.select('#articleBodyContents'):
    tag_text = tag_text.text.split('\n')[-3]
```

```python
tag_text
```

```python
from collections import Counter
word_count = Counter(tag_text.split(' '))
```

```python
word_count
```

```python
word_count.most_common()
```
